```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/winequality-red.csv', sep=';')
df.head()
```

```python
df.isnull().sum()
```

```python
feature = df.drop('quality', axis=1)
target = df['quality']
```

```python
lr = linear_model.LinearRegression()
lr.fit(feature, target)
%matplotlib inline
plt.xlabel('pH')
plt.ylabel('quality')
plt.title('Red Wine')
plt.scatter(df['pH'], df['quality'], color='blue', marker='+')
```

```python
lr.predict([[6.5, 0.8, 0.06, 22.5, 0.005, 12, 25, 1.0009, 3.05, 0.40, 10.8]])
```
