```python
import pandas as pd
df=pd.read_csv("/content/titanic - titanic.csv")
df.head(2)
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head(2)
```

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
inputs = pd.get_dummies(inputs, columns=['Gender'], drop_first=True, dtype=bool)
inputs.head(2)
```

```python
df.isnull().sum()
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs['Gender_male'] = inputs['Gender_male'].astype(int)
```

```python
inputs.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_test[0:3]
```

```python
model.predict(X_test[0:3])
```

```python
model.predict_proba(X_test[:3])
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),X_train, y_train, cv=5)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
print(f"Recall: {recall:.4f}")
print(f"Precision: {precision:.4f}")
```

```python
sample_input = pd.DataFrame([[3, 25, 10, 1]], columns=inputs.columns)
predicted_outcome = model.predict(sample_input)
outcome_text = "Survived" if predicted_outcome[0] == 1 else "Not survived"
print(f"The predicted outcome for the sample input is: {predicted_outcome[0]} ({outcome_text})")
```
