```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/email_dataset.csv")
df.head(2)
```

```python
inputs = df.drop('Label',axis='columns')
target = df.Label
```

```python
inputs = inputs.drop('Subject', axis='columns')
```

```python
inputs.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
inputs_encoded = inputs.copy()
for col in ['Contains_Link', 'Contains_Attachment', 'Urgent_Words', 'From_Trusted_Domain']:
    inputs_encoded[col] = le.fit_transform(inputs_encoded[col])

display(inputs_encoded.head(2))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_encoded.drop('Email_ID', axis='columns'), target, test_size=0.3, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test[:3])
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(GaussianNB(),X_train, y_train, cv=5)
print("Cross-validation scores:", scores)
print("Average accuracy:", np.mean(scores))
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred, pos_label='Phishing')
precision = precision_score(y_test, y_pred, pos_label='Phishing')
print(f"Accuracy: {accuracy:.4f}")
print(f"Recall: {recall:.4f}")
print(f"Precision: {precision:.4f}")
```
