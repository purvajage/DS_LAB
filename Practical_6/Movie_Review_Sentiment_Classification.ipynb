```python
import pandas as pd
df=pd.read_csv("/content/reviews.csv")
df.head(2)
```

```python
df.groupby('sentiment').describe()
```

```python
df['negative']=df['sentiment'].apply(lambda x: 1 if x=='negative' else 0)
df.head(2)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.review,df.negative)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
reviews = [
    'This movie was fantastic and I loved it!',
    'This movie was boring and I hated it.'
]
reviews_count = v.transform(reviews)
model.predict(reviews_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(reviews)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, classification_report
y_pred = model.predict(X_test_count)
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nPrecision Score:")
print(precision_score(y_test, y_pred))
print("\nRecall Score:")
print(recall_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
new_review = ["The cinematography was stunning, but the plot was dull"]
prediction = clf.predict(new_review)
print(f"The prediction for the review is: {prediction[0]}")
```
